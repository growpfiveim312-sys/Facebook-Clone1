```python
from ultralytics import YOLO

# Load trained model
model = YOLO("/content/runs/detect/train/weights/best.pt")

# (Optional) check class names to confirm coral class ID
print("Classes:", model.names)

# Set your coral class ID here (change if needed)
CORAL_CLASS_ID = 0

# Run inference on all images in folder
results = model(
    "/content/test/",   # folder of images
    conf=0.15,          # lower = more detections
    imgsz=640,
    save=True           # saves images with boxes
)

# Process results
for r in results:
    coral_detected = False

    for box in r.boxes:
        cls_id = int(box.cls[0])
        conf = float(box.conf[0])

        if cls_id == CORAL_CLASS_ID:
            coral_detected = True
            print(f"Detected coral in {r.path} (conf: {conf:.2f})")

    if not coral_detected:
        print(f"No coral detected in {r.path}")
```
